```python
import pandas as pd

def string_mean(series):
  """ Pandas 는 데이터프레임의 평균을 내는 내재 함수가 있지만, 데이터 전처리를 0으로 하여금
  문장이 긴 분절체와 단일 문장 데이터를 모두 수용함으로 빈공간도 모두 포함한다면 옳은 평균이 아니다"""
  sum = series.sum()
  length = len(series)
  temp_freq = 0

  if (series == 0).any():
    temp_freq = series.value_counts()
    temp_freq = temp_freq[0]    

  # 해당 열 [: 문장 분절체를 포함한 모든 확실도] 의 합을 열 길이 - 공백 데이터 수로 나눈다.
  return sum / (length - temp_freq)

def threshold(length, threshold_min = 30, threshold_max =  100, min_penalty = 0.5, max_penalty = 0.75):
  """ length 는 분절되지 않은 문장의 전체 길이를 말한다. 
  최소길이미만의 경우 min_penaly를 적용, 최대 길이 이상의 경우 max_penaly를 적용한다
  최소 길이 미만의 경우 감성분석의 정확도가 낮아 0.5로 제한하고, 임계 길이 이상의 경우 마침표로 분절하는 본 코드외 에
  문장이 길어질수록 전체적인 평균 수치가 결정되기 어렵기 때문에 그 수치를 제한하는 것에 목적이 있다."""

  if length >= threshold_max:
    return max_penalty
  elif length <= threshold_min:
    return min_penalty
  else:
    return 1.0

def evaluation(score):
  if -1 <= score and score < -0.75:
    return '매우 부정적'
  elif -0.75 <= score and score < -0.5:
    return '부정적'
  elif -0.5 <= score and score < -0.25:
    return '대체로 부정적'
  elif -0.25 <= score and score <= 0.25:
    return '미적지근함'
  elif 0.25 < score and score <= 0.5:
    return '대체로 긍정적'
  elif 0.5 < score and score <= 0.75:
    return '긍정적'
  elif 0.75 < score and score <= 1.0:
    return '매우 긍정적'

def sentence_tokenizer(string):

  string = string.replace("..", ".")

  if string.endswith('.'):
    return string[:-1]
  else:
    return string

def preprocessing(data_list):
  review = pd.Series(data_list)
  length_of_review = review.apply(len)
  review = review.apply(sentence_tokenizer)
  review = review.str.split('.', expand=True).fillna(0)

  return review, length_of_review
```
